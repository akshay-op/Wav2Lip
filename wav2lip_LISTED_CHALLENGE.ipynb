```python
from pytube import YouTube
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
import os


# video_url = input("Enter the youtube video url : ")
video_url = "https://www.youtube.com/watch?v=YMuuEv37s0o"
download_folder = "/content/sample_data/"

# Time range to trim (in seconds)

start_time = int(input(" Enter start time to trim :"))
end_time = int(input("Enter end time to trim :"))

yt = YouTube(video_url)
# stream = yt.streams.get_highest_resolution()
stream = yt.streams.filter(res="720p").first()
video_path = stream.download(output_path=download_folder)
output_filename = "input_video.mp4"
ffmpeg_extract_subclip(video_path, start_time, end_time, targetname=output_filename)
os.rename(output_filename, os.path.join(download_folder, output_filename))

print("Download and trimming completed!")

```
